```python
file = open("adventofcode_day2.txt","r")
lines = file.read().strip().splitlines()
```

```python
max(len(l) for l in lines)
```

```python
c = [0, 0]
for line in lines:
    hasTwo = False
    hasThree = False
    for letter in line:
        if not hasTwo and line.count(letter) == 2:
            c[0] += 1
            hasTwo = True
        if not hasThree and line.count(letter) == 3:
            c[1] += 1
            hasThree = True
        if hasTwo and hasThree:
            break
print(c)
```
